```python
from numpy import nan
import pandas as pd
s=pd.Series([1,2,3,4,5,6,7,8,None,nan])
print(s.describe())
print(s.map(lambda x:x*2))
print(s.apply(lambda x:x%2))
print(s.sort_values())
print(s.drop(1))
print(s.isnull())
print(s.notnull())
print(s.fillna(2))
print(s.dropna())
print(s.iloc[3])
print(s.iloc[-2])
print(s.loc[6])
print(s[s>=3])
print(s.sum())
print(s.cumsum())
print(s.iloc[3])
print(s.aggregate(['sum','mean','std']))
```
