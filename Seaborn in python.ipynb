```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
iris.head()

# 1. Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x="petal_length",y="petal_width",data=iris,hue="species")
plt.title("scatterplot of petal length vs petal width")
plt.show()

# 2. Line Plot
plt.figure(figsize=(6,4))
sns.lineplot(x="sepal_length", y="sepal_width", data=iris, ci=None)
plt.title("Line Plot - sepal length vs sepal width")
plt.show()

# 3. Bar Plot
plt.figure(figsize=(6,4))
sns.barplot(x="sepal_length", y="sepal_width", data=iris, hue="species")
plt.show()

# 4. Count Plot
plt.figure(figsize=(5,4))
sns.countplot( x="sepal_length", data=iris, hue="species")
plt.title("Count Plot - sepal length vs sepal width")
plt.show()

# 5. Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x="petal_length", y="petal_width", data=iris, hue="species")
plt.title("Box Plot - petal length vs petal width")
plt.show()
iris.head()

# 6. Violin Plot
plt.figure(figsize=(6,4))
sns.violinplot(x="sepal_length", y="sepal_width", data=iris, hue="species", split=True)
plt.title("Violin Plot - sepal length vs sepal width")
plt.show()

# 7. Heatmap (Correlation Matrix)
plt.figure(figsize=(6,4))
corr = iris.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# 8. Pairplot
sns.pairplot(iris, hue="species")
plt.suptitle("Pairplot of iris Dataset",y=1.02)
plt.show()

# 9. Distribution Plot (Histogram + KDE)
plt.figure(figsize=(6,4))
sns.histplot(iris["petal_length"], kde=True, bins=20)
plt.title("Distribution of petal_length")
plt.show()

# 10. Regression Plot
plt.figure(figsize=(6,4))
sns.regplot(x="petal_length", y="petal_width", data=iris)
plt.title("Regression Plot ")
plt.show()
```
