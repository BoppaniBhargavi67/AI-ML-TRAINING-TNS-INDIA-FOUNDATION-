```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

# Dataset
df = pd.DataFrame({
    "Email": [
        "Congratulations! You won a free iPhone",
        "Meeting is at 10 AM tomorrow",
        "Claim your cash prize now",
        "Project report submitted successfully",
        "Limited offer! Buy now",
        "Please attend the team meeting",
        "You have won a lottery",
        "Assignment deadline is tomorrow",
        "Free vacation package waiting for you",
        "Click here to claim your reward"
    ],
    "Label": [
        "Spam",
        "Not Spam",
        "Spam",
        "Not Spam",
        "Spam",
        "Not Spam",
        "Spam",
        "Not Spam",
        "Spam",
        "Spam"
    ]
})

# Independent Variable
X = df["Email"]

# Dependent Variable
y = df["Label"]

# Convert text into numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X)

# Create Model
model = LogisticRegression()

# Train Model
model.fit(X, y)

# Prediction
new_email = ["Congratulations! Claim your free cash prize"]

new_email = vectorizer.transform(new_email)

prediction = model.predict(new_email)

print("Prediction:", prediction[0])

# Visualization
df["Label"].value_counts().plot(kind="bar")

plt.title("Spam vs Not Spam Emails")
plt.xlabel("Email Type")
plt.ylabel("Count")
plt.grid(axis="y")

plt.show()
```
