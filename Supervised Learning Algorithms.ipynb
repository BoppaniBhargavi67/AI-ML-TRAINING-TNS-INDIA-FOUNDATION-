**Linear** **Regression**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
df = pd.DataFrame({
    "Temperature": [20, 22, 24, 26, 28, 30, 32],
    "IceCreamSales": [100, 120, 140, 160, 180, 200, 220]
})

# Independent Variable
X = df[["Temperature"]]

# Dependent Variable
y = df["IceCreamSales"]

# Create Model
model = LinearRegression()

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[35]])

print("Predicted Ice Cream Sales at 35°C:", prediction[0])

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)

# Visualization
plt.figure(figsize=(6,4))

# Scatter plot
plt.scatter(df["Temperature"], df["IceCreamSales"], label="Actual Data")

# Regression line
plt.plot(df["Temperature"], model.predict(X), label="Regression Line")

plt.title("Temperature vs Ice Cream Sales")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales")
plt.legend()
plt.grid(True)

plt.show()
```

**Multiple Linear Regression**

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
df = pd.DataFrame({
    "Rainfall": [500, 550, 600, 650, 700, 750, 800],
    "Fertilizer": [40, 45, 50, 55, 60, 65, 70],
    "Crop_Yield": [2.5, 2.8, 3.2, 3.6, 4.0, 4.4, 4.8]
})

# Independent Variables
X = df[["Rainfall", "Fertilizer"]]

# Dependent Variable
y = df["Crop_Yield"]

# Create Model
model = LinearRegression()

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[720, 62]])

print("Predicted Crop Yield:", prediction[0], "tons/hectare")

print("\nCoefficients:")
print("Rainfall:", model.coef_[0])
print("Fertilizer:", model.coef_[1])

print("Intercept:", model.intercept_)

# Rainfall vs Crop Yield
plt.figure(figsize=(6,4))
plt.scatter(df["Rainfall"], df["Crop_Yield"])
plt.title("Rainfall vs Crop Yield")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Crop Yield (tons/hectare)")
plt.grid(True)
plt.show()

# Fertilizer vs Crop Yield
plt.figure(figsize=(6,4))
plt.scatter(df["Fertilizer"], df["Crop_Yield"])
plt.title("Fertilizer vs Crop Yield")
plt.xlabel("Fertilizer Used (kg)")
plt.ylabel("Crop Yield (tons/hectare)")
plt.grid(True)
plt.show()
```

**Polynomial** **Regression**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Dataset
df = pd.DataFrame({
    "Days": [1, 2, 3, 4, 5, 6, 7],
    "Plant_Height": [2, 4, 7, 11, 16, 22, 29]   # Height in cm
})

# Independent Variable
X = df[["Days"]]

# Dependent Variable
y = df["Plant_Height"]

# Convert to Polynomial Features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Create Model
model = LinearRegression()

# Train Model
model.fit(X_poly, y)

# Prediction
prediction = model.predict(poly.transform([[8]]))

print("Predicted Plant Height on Day 8:", prediction[0], "cm")

print("\nCoefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Visualization
plt.figure(figsize=(6,4))

# Scatter Plot
plt.scatter(df["Days"], df["Plant_Height"], label="Actual Data")

# Polynomial Curve
X_range = np.linspace(1, 8, 100).reshape(-1,1)
plt.plot(
    X_range,
    model.predict(poly.transform(X_range)),
    label="Polynomial Regression Curve"
)

plt.title("Days vs Plant Height")
plt.xlabel("Days")
plt.ylabel("Plant Height (cm)")
plt.legend()
plt.grid(True)

plt.show()
```

**Decision Tree**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Dataset
df = pd.DataFrame({
    "Age": [18, 22, 25, 30, 35, 40, 45, 50],
    "Monthly_Income": [15000, 20000, 25000, 35000, 45000, 55000, 65000, 75000],
    "Join_Gym": ["No", "No", "Yes", "Yes", "Yes", "No", "No", "No"]
})

# Independent Variables
X = df[["Age", "Monthly_Income"]]

# Dependent Variable
y = df["Join_Gym"]

# Create Decision Tree Model
model = DecisionTreeClassifier(random_state=42)

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[28, 30000]])

print("Gym Membership Prediction:", prediction[0])

# Visualize Decision Tree
plt.figure(figsize=(10,6))
plot_tree(
    model,
    feature_names=["Age", "Monthly Income"],
    class_names=["No", "Yes"],
    filled=True
)

plt.title("Decision Tree - Gym Membership Prediction")
plt.show()
```

**Random Forest**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Dataset
df = pd.DataFrame({
    "AC_Hours": [1, 2, 3, 4, 5, 6, 7],
    "Fan_Hours": [4, 5, 5, 6, 7, 8, 9],
    "Electricity_Units": [6, 9, 13, 18, 24, 31, 39]
})

# Independent Variables
X = df[["AC_Hours", "Fan_Hours"]]

# Dependent Variable
y = df["Electricity_Units"]

# Create Model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[5, 7]])

print("Predicted Electricity Consumption:", prediction[0], "units")

# Feature Importance
print("\nFeature Importance:")
print("AC Hours:", model.feature_importances_[0])
print("Fan Hours:", model.feature_importances_[1])

# Visualization
plt.figure(figsize=(6,4))
plt.scatter(df["AC_Hours"], df["Electricity_Units"], label="Actual Data")

# Predicted values
pred = model.predict(X)
plt.scatter(df["AC_Hours"], pred, marker="x", label="Predicted")

plt.title("AC Hours vs Electricity Consumption")
plt.xlabel("AC Usage (Hours)")
plt.ylabel("Electricity Units")
plt.legend()
plt.grid(True)
plt.show()
```

**Support Vector** **Regression**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

# Dataset
df = pd.DataFrame({
    "Exercise_Time": [10, 20, 30, 40, 50, 60, 70],
    "Calories_Burned": [60, 110, 170, 240, 320, 410, 500]
})

# Independent Variable
X = df[["Exercise_Time"]].values

# Dependent Variable
y = df["Calories_Burned"].values

# Feature Scaling
sc_X = StandardScaler()
sc_y = StandardScaler()

X_scaled = sc_X.fit_transform(X)
y_scaled = sc_y.fit_transform(y.reshape(-1, 1)).ravel()

# Create SVR Model
model = SVR(kernel="rbf")

# Train Model
model.fit(X_scaled, y_scaled)

# Prediction
prediction = model.predict(sc_X.transform([[45]]))
prediction = sc_y.inverse_transform(prediction.reshape(-1, 1))

print("Predicted Calories Burned:", prediction[0][0])

# Visualization
plt.figure(figsize=(6,4))

# Actual Data
plt.scatter(df["Exercise_Time"], df["Calories_Burned"], label="Actual Data")

# SVR Curve
X_range = np.linspace(10, 70, 100).reshape(-1,1)
y_pred = model.predict(sc_X.transform(X_range))
y_pred = sc_y.inverse_transform(y_pred.reshape(-1,1))

plt.plot(X_range, y_pred, label="SVR Curve")

plt.title("Exercise Time vs Calories Burned")
plt.xlabel("Exercise Time (minutes)")
plt.ylabel("Calories Burned")
plt.legend()
plt.grid(True)

plt.show()
```

**Logistic Regression**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Dataset
df = pd.DataFrame({
    "Attendance": [60, 65, 70, 75, 80, 85, 90, 95],
    "Exam_Score": [50, 55, 60, 68, 72, 80, 88, 95],
    "Scholarship": ["No", "No", "No", "Yes", "Yes", "Yes", "Yes", "Yes"]
})

# Independent Variables
X = df[["Attendance", "Exam_Score"]]

# Dependent Variable
y = df["Scholarship"]

# Create Model
model = LogisticRegression()

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[78, 70]])

print("Scholarship Prediction:", prediction[0])

# Visualization
plt.figure(figsize=(6,4))

colors = {"Yes": "green", "No": "red"}

for label in colors:
    subset = df[df["Scholarship"] == label]
    plt.scatter(
        subset["Attendance"],
        subset["Exam_Score"],
        label=label
    )

# New Student
plt.scatter(78, 70, marker="*", s=200, label="New Student")

plt.title("Logistic Regression - Scholarship Prediction")
plt.xlabel("Attendance (%)")
plt.ylabel("Exam Score")
plt.legend()
plt.grid(True)

plt.show()
```

**K-Nearest Neighbors (KNN)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Dataset
df = pd.DataFrame({
    "Age": [18, 22, 25, 28, 32, 36, 40, 45],
    "Income": [15000, 20000, 25000, 30000, 45000, 50000, 60000, 70000],
    "Buy_Mobile": ["No", "No", "Yes", "Yes", "Yes", "Yes", "No", "No"]
})

# Independent Variables
X = df[["Age", "Income"]]

# Dependent Variable
y = df["Buy_Mobile"]

# Create KNN Model
model = KNeighborsClassifier(n_neighbors=3)

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[30, 35000]])

print("Prediction:", prediction[0])

# Visualization
plt.figure(figsize=(6,4))

# Color mapping
colors = {"Yes": "green", "No": "red"}

for label in colors:
    subset = df[df["Buy_Mobile"] == label]
    plt.scatter(
        subset["Age"],
        subset["Income"],
        label=label
    )

# New customer
plt.scatter(30, 35000, marker="*", s=200, label="New Customer")

plt.title("KNN - Mobile Purchase Prediction")
plt.xlabel("Age")
plt.ylabel("Monthly Income")
plt.legend()
plt.grid(True)

plt.show()
```

**unique Naive Bayes Classification**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB

# Dataset
df = pd.DataFrame({
    "IMDb_Rating": [5.2, 5.8, 6.3, 6.8, 7.2, 7.8, 8.4, 9.0],
    "Duration": [90, 95, 100, 110, 120, 130, 140, 150],
    "Watch_Movie": ["No", "No", "No", "Yes", "Yes", "Yes", "Yes", "Yes"]
})

# Independent Variables
X = df[["IMDb_Rating", "Duration"]]

# Dependent Variable
y = df["Watch_Movie"]

# Create Model
model = GaussianNB()

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[7.0, 115]])

print("Movie Recommendation:", prediction[0])

# Visualization
plt.figure(figsize=(6,4))

colors = {"Yes": "green", "No": "red"}

for label in colors:
    subset = df[df["Watch_Movie"] == label]
    plt.scatter(
        subset["IMDb_Rating"],
        subset["Duration"],
        label=label
    )

# New Movie
plt.scatter(7.0, 115, marker="*", s=200, label="New Movie")

plt.title("Naive Bayes - Movie Recommendation")
plt.xlabel("IMDb Rating")
plt.ylabel("Duration (minutes)")
plt.legend()
plt.grid(True)

plt.show()
```
