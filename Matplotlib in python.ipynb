**Matplotlib**

```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

subjects = ["Python", "Java", "C", "C++"]
marks = [85, 78, 90, 80]

hist_data = [10, 20, 20, 30, 30, 30, 40, 50]

pie_labels = ["Python", "Java", "C"]
pie_sizes = [50, 30, 20]

y2 = [5, 15, 25, 35, 45]

# Create Figure
plt.figure(figsize=(15, 10))

# 1. Line Plot
plt.subplot(3, 3, 1)
plt.plot(x, y, marker='o')
plt.title("Line Plot")

# 2. Bar Graph
plt.subplot(3, 3, 2)
plt.bar(subjects, marks)
plt.title("Bar Graph")

# 3. Histogram
plt.subplot(3, 3, 3)
plt.hist(hist_data, bins=5)
plt.title("Histogram")

# 4. Scatter Plot
plt.subplot(3, 3, 4)
plt.scatter(x, y)
plt.title("Scatter Plot")

# 5. Pie Chart
plt.subplot(3, 3, 5)
plt.pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%')
plt.title("Pie Chart")

# 6. Multiple Lines
plt.subplot(3, 3, 6)
plt.plot(x, y, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.legend()
plt.title("Multiple Lines")

# 7. Customized Graph
plt.subplot(3, 3, 7)
plt.plot(x, y, color='red', linestyle='--', marker='o', linewidth=2)
plt.title("Customized Graph")

# 8. Horizontal Bar Graph
plt.subplot(3, 3, 8)
plt.barh(subjects, marks)
plt.title("Horizontal Bar")

# 9. Box Plot
plt.subplot(3, 3, 9)
plt.boxplot(hist_data)
plt.title("Box Plot")

plt.tight_layout()
plt.show()
```
