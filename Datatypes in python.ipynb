**LIST**

```python
student=["bhanu","sunil","sindhu","madhu"]
print(student)
student.append("bhona")
print(student)
student.insert(1,"bhargavi")
print(student)
student.remove("madhu")
print(student)
student.pop(2)
print(student)
student.sort()
print(student)
student.reverse()
print(student)
l=len(student)
print(l)
c=["bhaaanu"]
c.count("a")
student.count("bhargavi")

```

**TUPLE**

```python
student=("bhanu","sunil","sindhu","madhu")
print(student)
l=len(student)
print(l)
print(student[2])
```

**SET**

```python
name1 = {"bhanu","sunil","sindhu","madhu"}
name2 = {"bhanu","sunil","bhona","bhargavi"}
print(name1)
print(name2)
name1.add('archu')
print(name1)
name2.remove('bhargavi')
print(name2)
unions = name1.union(name2)
print(unions)
inter = name1.intersection(name2)
print(inter)
dif = name1.difference(name2)
print(dif)
name1.clear()
print(name1)
```
