```python
import pandas as pd
df=pd.read_csv("weight-height.csv")
df
```

```python
df.describe()
```

```python
#Detect outliers(height) using IQR
Q1=df.Height.quantile(0.25)
Q3=df.Height.quantile(0.75)
Q1,Q3
```

```python
#finding IQR
IQR=Q3-Q1
IQR
```

```python
#finding lower,upper limits
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
#Here are the outliers

df[(df.Height<lower_limit)|(df.Height>upper_limit)]
```

```python
# replacing outliers with mean
mean_h = df['Height'].mean()
df['Height'] = df['Height'].apply(
    lambda x: mean_h if x <lower_limit or x > upper_limit else x
)

```

```python
#Detect outliers(height) using IQR
Q1=df.Weight.quantile(0.25)
Q3=df.Weight.quantile(0.75)
Q1,Q3
```

```python
#finding IQR
IQR=Q3-Q1
IQR
```

```python
#finding lower,upper limits
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
#Here are the outliers

df[(df.Weight<lower_limit)|(df.Weight>upper_limit)]
```

```python
# replacing outliers with mean
mean_h = df['Weight'].mean()
df['Weight'] = df['Weight'].apply(
    lambda x: mean_h if x <lower_limit or x > upper_limit else x
)
```
