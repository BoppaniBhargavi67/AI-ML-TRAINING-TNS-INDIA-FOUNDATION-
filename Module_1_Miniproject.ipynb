**Netflix Data Visualization Project**

```python
#importing dataset from kaggle
from google.colab import files
uploaded=files.upload()

#importing librarires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#reqading the uploaded netflix file
df = pd.read_csv('netflix_titles.csv')
df.head()
df.info()

```

```python
# Which type of content is more available on Netflix?
content_count = df['type'].value_counts()
print(content_count)

#visualizing top content in netflix
plt.figure(figsize=(10,6))
sns.countplot(x='type', data=df)
plt.title('Movies vs TV Shows on Netflix')
plt.show()

#Which countries contribute the most content to Netflix?
country_count = df['country'].value_counts().head(5)
print(country_count)

#visualizing tha data(countries mostly contribute content to netfilx)
plt.figure(figsize=(10,6))
sns.barplot(x=country_count.index, y=country_count.values)
plt.title('Top 5 Countries with Most Content')
plt.xlabel('Country')
plt.ylabel('Number of Content')
plt.show

#Which directors have the most titles on Netflix?
director_count = df['director'].value_counts().head(5)
print(director_count)

#visualizing tha data(directors mostly have titles on netfilx)
plt.figure(figsize=(10,6))
sns.barplot(x=director_count.index, y=director_count.values)
plt.title('Top 5 Directors with Most Titles')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.show


```
