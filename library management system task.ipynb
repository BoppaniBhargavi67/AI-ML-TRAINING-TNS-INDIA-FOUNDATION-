```python
import pandas as pd
import numpy as np

# Create Dataset
data = {
    "Book_ID": [None, 102, 103, 104, 105],
    "Book_Name": ["Python Basics", "Data Science", "Machine Learning",
                  "Deep Learning", "History of India"],
    "Category": ["Programming", "Programming", "AI", "AI", "History"],
    "Borrow_Count": [120, 150, 180, 160, 90],
    "Rating": [4.5, 4.8, 4.9, 4.7, 4.2]
}

df = pd.DataFrame(data)

# Inspect Dataset
print("Dataset:")
print(df)

# Handle Missing Values
print(df.isnull().sum())

# Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
# Find Books Borrowed More Than Average
avg_borrow = df["Borrow_Count"].mean()
above_avg = df[df["Borrow_Count"] > avg_borrow]

print("\nBooks Borrowed More Than Average:")
print(above_avg[["Book_Name", "Borrow_Count"]])

# Sort by Borrow Count
sorted_books = df.sort_values(by="Borrow_Count", ascending=False)

print("\nSorted by Borrow Count:")
print(sorted_books[["Book_Name", "Borrow_Count"]])
```

```python
#Category-wise Average Ratings
print("\nCategory-wise Average Ratings:")
print(df.groupby("Category")["Rating"].mean())

# Create Popularity Index
df["Popularity_Index"] = df["Borrow_Count"] * df["Rating"]

print("\nPopularity Index:")
print(df[["Book_Name", "Popularity_Index"]])

# Apply NumPy Functions
print("\nMaximum Borrow Count:", np.max(df["Borrow_Count"]))
print("Minimum Borrow Count:", np.min(df["Borrow_Count"]))
print("Average Rating:", np.mean(df["Rating"]))

# Most Borrowed Book
most_borrowed = df.loc[df["Borrow_Count"].idxmax()]
print("\nMost Borrowed Book:")
print(most_borrowed["Book_Name"])

# Generate Report
print("\n----- LIBRARY REPORT -----")
print("Total Books:", len(df))
print("Average Borrow Count:", avg_borrow)
print("Most Borrowed Book:", most_borrowed["Book_Name"])
print("--------------------------")
```
