Mini Project:

Create a Student Report Card Program using:

1.List → store student names

2.Set → store subjects

3.Tuple → store student information (name, age)

4.Dictionary → store subject-wise marks

Requirements:

1.Ask the user to enter:

*Student name

*Age

*Number of subjects

2.For each subject:

*Enter subject name

*Enter marks

3.Store:

*Name in a list

*Subjects in a set

*Student info in a tuple

*Marks in a dictionary

4.Print a report card showing:

*Name

*Age

*Subjects

*Marks

*Total marks

*Average marks

```python
student_name = [input("Enter student name: ")]
student_age = {int(input("Enter student age: "))}
no_of_subjects = (int(input("Enter no of subjects: ")),)

subject_wise_marks = {}

for i in range(no_of_subjects[0]):
    subject = input("Enter subject name: ")
    marks = int(input("Enter marks: "))
    subject_wise_marks[subject] = marks

print("\nStudent Name:", student_name)
print("Student Age:", student_age)
print("Number of Subjects:", no_of_subjects)
print("Subject-wise Marks:", subject_wise_marks)

total = sum(subject_wise_marks.values())
average = total / no_of_subjects[0]

print("Total Marks =", total)
print("Average Marks =", average)2



```
